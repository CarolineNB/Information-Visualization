```
from google.colab import drive
import pandas as pd
```

```
df = pd.read_csv("https://raw.githubusercontent.com/CarolineNB/nyc-collision-dataset-csv/master/aiddata-countries-only%20-%20aiddata-countries-only.csv")
df
```

```
donors = df.groupby('donor')['commitment_amount_usd_constant'].sum()
recipients = df.groupby('recipient')['commitment_amount_usd_constant'].sum()
```

```
result = donors - recipients
result = result.sort_values()
result
```

```
rows_Nan = result.isnull()
were_nan = recipients[rows_Nan] * -1
```

```
final_results = result.append(were_nan).dropna()
```

```
ordered_results = final_results.sort_values(ascending=False)
```
